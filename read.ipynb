```python
import openpyxl
```

```python
workbook = openpyxl.load_workbook(filename = '截至2020年12月16日B站电竞行业播放数前100位的视频情况.xlsx')
print(workbook.sheetnames)
```

```python
sheet = workbook['数据']
print(sheet)
print(dir(sheet))
```

```python
print(sheet.rows)
print(sheet.columns)
```

```python
sheet.min_row
sheet.max_column
```

```python
for row in sheet.rows:
    print(
        '视频名称：%s，UP主：%s，播放量：%s，弹幕数：%s，投稿时间：%s'
        % (row[0].value, row[1].value, row[2].value, row[3].value, row[4].value)
    )
```

```python
for row in sheet.iter_rows(min_row=3, max_row=102):
    # 获取第三行到第102行的数据
    '视频名称：%s，UP主：%s，播放量：%s，弹幕数：%s，投稿数：%s' % (
        row[0].value,
        row[1].value,
        row[2].value,
        row[3].value,
        row[4].value,
    )
```
