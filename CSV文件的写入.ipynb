```python
with open("截至2020年12月16日B站电竞行业播放数前100位的视频情况.csv", encoding="utf-8-sig", newline='') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        print(row)
```
