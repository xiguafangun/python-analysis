```python
import numpy as np

# 创建一个一维数组
a = np.array([0, 1, 2, 3, 4, 5])
# 创建一个二维数组
b = np.array(
    [
        [11, 12, 13, 14, 15],
        [21, 22, 23, 24, 25],
        [31, 32, 33, 34, 35],
        [41, 42, 43, 44, 45],
        [41, 42, 43, 44, 45],
    ]
)

print("a 的维数：", a.ndim)
print("a 的元素数量：", a.size)
print("a 的形状：", a.shape)


print("b 的维数：", b.ndim)
print("b 的元素数量：", b.size)
print("b 的形状：", b.shape)
```

```python
print(np.zeros((3, 4))) # 创建一个形状 3 x 4 所有元素都为 0 的数组
print(np.ones( (2, 3), dtype=np.int16 )) # 创建一个形状 2 x 3
                                         # 所有元素都为 1
                                         # 类型为 int16 的数组
print(np.empty((4, 35))) # 创始数组时不进行初始化，里面的元素可能为任何值
```

```python
# 针对一维数组切片
print(a[2:4])
print(a[2:])
print(a[:4])
print(a[:])
print(a[::2])
```

```python
# 针对二维数组切片
print(b[:2])
print(b[2:4, 2:4])
print(b[:, 2:4])
print(b[:, 2])
```

```python
print(a.reshape(2, 3))
```

```python
print(b.ravel())
```

```python
print(b.reshape(-1))
```
