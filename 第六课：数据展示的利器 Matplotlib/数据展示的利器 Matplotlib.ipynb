```python
# 2012-2018年全球电竞市场营收情况

data = {
    "2012": 90,
    "2013": 180,
    "2014": 140,
    "2015": 240,
    "2016": 350,
    "2017": 460,
    "2018": 600,
}

plt.title("2012-2018年全球电竞市场营收情况")
plt.xlabel("职业")
plt.ylabel("营收(百万美元)")

plt.plot(data.keys(), data.values())

plt.show()
```
